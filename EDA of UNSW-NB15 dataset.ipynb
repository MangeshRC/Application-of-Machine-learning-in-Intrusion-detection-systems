```python
# importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
'''
    Was getting a warning with datatype while reading these large csv files because pandas have to read 
    complete file before deciding the datatype for each column. Referred below link to resolve the issue.
'''
# https://stackoverflow.com/questions/24251219/pandas-read-csv-low-memory-and-dtype-options

data1 = pd.read_csv("UNSW-NB15_1.csv",low_memory=False)
data2 = pd.read_csv("UNSW-NB15_2.csv",low_memory=False)
data3 = pd.read_csv("UNSW-NB15_3.csv",low_memory=False)
data4 = pd.read_csv("UNSW-NB15_4.csv",low_memory=False)
```

```python
'''
We can see the shape and the number of datapoints each csv file has.
Each has 49 columns and the number of datapoints as printed below.
'''

print("UNSW-NB15_1.csv shape : ",data1.shape)
print("UNSW-NB15_2.csv shape : ",data2.shape)
print("UNSW-NB15_3.csv shape : ",data3.shape)
print("UNSW-NB15_4.csv shape : ",data4.shape)
```

```python
# there are no column names in any of the above read files.
# Thus column names have to be added manually

data1.head()
```

```python
'''
    "UNSW-NB15_features.csv" file contains all the feature names which will be added as column names in above read files.
    There are 4 columns in this file
    1. Sr. No
    2. Feature name
    3. Feature type
    4. Feature Description

'''

class_labels = pd.read_csv("UNSW-NB15_features.csv")

class_labels.head()
```

```python
'''
We need to merge all 4 files but before merging all the column names should be added
Since the dataframes will be merged based on column names.

This cell adds the column names and merge all 4 dataframe to create complete dataset
'''

col = class_labels['Name'].values   # Getting Feature names


# Adding columns to all dataframes
data1 = data1.set_axis(col,axis=1)
data2 = data2.set_axis(col,axis=1)
data3 = data3.set_axis(col,axis=1)
data4 = data4.set_axis(col,axis=1)


# Merging all dataframes
data_df = pd.concat([data1,data2,data3,data4],axis=0,ignore_index=True)

# Saving the final dataframe on disk
data_df.to_csv("final_data.csv",index=False)

data_df.head()
```

## Cleaning Data

### a. Whitespace in One of the Column Names

```python
'''
    It was observed in EDA that ct_src_ltm column name was having whitespace in it's name.
    This needs to be removed in order to proceed with EDA.
'''

data_df = pd.read_csv("final_data.csv",low_memory=False)

data_df.columns[-6]
```

```python
# Got this list from previous cell and removed whitespace to update column names again
col = ['srcip', 'sport', 'dstip', 'dsport', 'proto', 'state', 'dur', 'sbytes',
       'dbytes', 'sttl', 'dttl', 'sloss', 'dloss', 'service', 'Sload', 'Dload',
       'Spkts', 'Dpkts', 'swin', 'dwin', 'stcpb', 'dtcpb', 'smeansz',
       'dmeansz', 'trans_depth', 'res_bdy_len', 'Sjit', 'Djit', 'Stime',
       'Ltime', 'Sintpkt', 'Dintpkt', 'tcprtt', 'synack', 'ackdat',
       'is_sm_ips_ports', 'ct_state_ttl', 'ct_flw_http_mthd', 'is_ftp_login',
       'ct_ftp_cmd', 'ct_srv_src', 'ct_srv_dst', 'ct_dst_ltm', 'ct_src_ltm',
       'ct_src_dport_ltm', 'ct_dst_sport_ltm', 'ct_dst_src_ltm', 'attack_cat',
       'Label']


data_df = data_df.set_axis(col,axis=1)
```

### b. Updating sport and dsport hex values

```python
'''
    1. During EDA it was found that sport and dsport columns had very few data points with hex numbers.
    2. These all points were belonging to normal network traffic.
    3. Since these points are very few and because of these points it was difficult to analyze other
        datapoints and column was loosing it's numerical nature, these rows should be removed.

'''
print("Shape of data before removal : ",data_df.shape)

# 1. sport column

# empty list to store non-numerical values
non_numerical = []

# loop to get all non numerical values
for i in data_df['sport'].values:
    try:
        int(i)
    except:
        if i not in non_numerical:
            non_numerical.append(i)
            
print("These are  non numerical values in sport column : ",non_numerical)


# removing non-numerical value rows
data_df = data_df[data_df.sport != '0x000b']
data_df = data_df[data_df.sport != '0x000c']
data_df = data_df[data_df.sport != '-']


# 2. dsport column

# empty list to store non-numerical values
non_numerical = []


# loop to get all non numerical values
for i in data_df['dsport'].values:
    try:
        int(i)
    except:
        if i not in non_numerical:
            non_numerical.append(i)
            
print("These are  non numerical values in dsport column : ",non_numerical)


# removing non-numerical value rows
data_df = data_df[data_df.dsport != '0xc0a8']
data_df = data_df[data_df.dsport != '-']
data_df = data_df[data_df.dsport != '0x20205321']
data_df = data_df[data_df.dsport != '0xcc09']

print("Shape of data after removal : ",data_df.shape)
```

<li> It can be seen from shape before and after removal that very few datapoints were removed. </li>

### c. Missing Values in ct_ftp_cmd feature 

```python
'''
While performing EDA it was observed that ct_ftp_cmd column has values 0 to 8
But instead of 7 there was a whitespace with which it was not possible to perform EDA
Hence in this cell the whitespace was removed and replaced with 7
'''


print("Unique Values in column : ",data_df['ct_ftp_cmd'].unique())

# These are the distict values in ct_ftp_cmd column. ( white space excluded)
temp = ['0', '1', '6', '2', '4', '8', '5', '3']

data_df['ct_ftp_cmd'] = data_df['ct_ftp_cmd'].apply(lambda x:x if x in temp else '7')

print("Unique Values in column after replacement : ",data_df['ct_ftp_cmd'].unique())
```

### d. Handling Null Values

```python
'''
This cell checks for any null values in any of the columns
'''

print(data_df.isna().any())
```

#### We can see there are null values in "ct_flw_http_mthd",  "is_ftp_login" and "attack_cat"  columns. These need to processed before proceeding with any EDA

#### d.1 Updating 'normal' as one catagory in attack_cat column for normal network traffic

```python
'''
attack_cat column lists types of attacks for Label column value 1
but for normal traffic, the values are null in 'attack_cat' column.
Hence these need to be updated to 'normal' to perform EDA
'''

# updating 'normal' where null is present
data_df['attack_cat'] = ['normal' if i[1]==0 else i[0] for i in data_df[['attack_cat','Label']].values]
```

```python
# Now after handling Null values in attac_cat column, let us try to remove and null values for other columns

print("Shape with NaN values    : ",data_df.shape)

print("Shape without NaN values : ",data_df.dropna().shape)
```

<li> We can see that just by removing null value rows we are loosing a large number of the datapoints in dataset </li>
<li> Hence let us analyze each column in detail, then only we will proceed with removal or substitution </li>

#### d.2 ct_flw_http_mthd Column

```python
'''
1. The first print statement shows 1348143 datapoints have null values for ct_flw_http_mthd
2. This is a huge number and it is not advisable to remove these many datapoints
3. Hence replacing null values with mean of all column values
'''

print(data_df['ct_flw_http_mthd'].isna().value_counts())

# replacing with mean value of all column elements
data_df['ct_flw_http_mthd'] = data_df['ct_flw_http_mthd'].fillna(data_df['ct_flw_http_mthd'].mean())
data_df['ct_flw_http_mthd'].value_counts()
```

<li> We can see now 1348143 datapoints were updated to have 0.234586 as 'ct_flw_http_mthd' value</li>

#### d.2 is_ftp_login Column

```python
'''
    1. The first print statement shows 1429877 datapoints have null values for 'is_ftp_login'
    2. This is a huge number and it is not advisable to remove these many datapoints
    3. Since this is a Flag to indicate whether ftp session used or not, the value corresponding to this column should be 
        either 1 or 0.
    4. It can be seen that there other values like 4, 2 which should not be there.
    5. Hence updating with 0.5 for all null and these exception values. We can update with any value since this 
        is going to be a categorical variable. Thus while encoding with OneHotEncoder this value will not have 
        an impact.
'''

print(data_df['is_ftp_login'].isna().value_counts())

print("\n\n")
print(data_df['is_ftp_login'].value_counts())

data_df['is_ftp_login'][(data_df.is_ftp_login != 0) & (data_df.is_ftp_login != 1)] = 0.5

print("\n\n")
print(data_df['is_ftp_login'].value_counts())
```

```python
# Reading processed data from disk

data_df = pd.read_csv("final_data.csv",low_memory=False)
```

#### Since our focus is to detect attack or non-attack, let us anaylyze the distribution of each class in dataset

```python
labels = data_df['Label'].values         # Getting label column from dataset

# Counting number of data points in each category
label_count = [data_df['Label'].value_counts()[0],data_df['Label'].value_counts()[1]]

# class labels
categ_label = ['Normal Traffic','Attack Traffic']


print("Normal Traffic Points : ",label_count[0],"\nAttack Traffic Points : ",label_count[1])

# Referred pie chart from here : https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html
plt.figure(figsize=(5,5))
plt.pie(label_count,labels=categ_label)
plt.title("Normal and Attack traffic")
plt.show()
```

####  It can be seen with above pie chart that there is a high imbalance in positive and negative data points

#### Now let us focus more on types of attacks. Now it is clear that attacks have lesser data points but let us see the distribution of each attack category ( normal traffic is excluded for this analysis )

```python
# Some whitespaces were found in class labels while performing EDA thus removing those
data_df['attack_cat'] = data_df.attack_cat.apply(lambda x:x.strip())


attack_cat = list(data_df['attack_cat'].value_counts().keys())[1:]    # All attack names excluding 'normal'
attack_count = list(data_df['attack_cat'].value_counts())[1:]         # All attack counts


print("Number of attack categories are : ",len(attack_cat))

print("Number of datapoints in each attack category are as follows: \n")

for i in range(len(attack_cat)):       # Loop to iterate through all attack categories and count those.
    print("{:20}".format(attack_cat[i]),attack_count[i])
        

plt.figure(figsize=(8,8))
plt.pie(attack_count,labels=attack_cat)
plt.title("Attack Categories")
plt.show()
```

#### From above class analysis following conclusions can be made:
<ul>
    <li> Non-attack points are in majority when compared to attack points </li>
    <li> Thus it is sure that dataset will be an imbalanced dataset for binary classification tasks </li>

## Conclusions based on performed EDA

<li> There is an imbalance in data both for binary classification, as well as attack classification. </li>

#### 3. Virtual Average packet size of a datapoint from Source to destination (vr_source_avg_pkt)
<br>


<li> We have Start record time (Stime), Last record time (Ltime),Source to destination transfer rate in bits/sec (Sload) and Source to destination packet count (Spkts). Based on these features, we can derive virtual source average packet size for a particular datapoint. </li>
<li> This can be calculated using formula ((Ltime - Stime)*Sload)/Spkts . </li>
<li> Since (Ltime - Stime) gives total time in seconds, Sload is data rate in bits/sec so (Ltime - Stime)*Sload gives us total data transferred in bits (assuming continuous transfer). Dividing it by Spkts will give us virtual average packet size for that particular datapoint. </li>

```python
# Above formula is implemented with below statement and stored in source_avg_pkt column

data_df['vr_source_avg_pkt'] = ((data_df['Ltime'].values-data_df['Stime'].values)*data_df['Sload'].values)/data_df['Spkts'].values
```

```python
feature = 'vr_source_avg_pkt'

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle("Class Labels vs {}".format(feature))

sns.violinplot(x="Label", y=feature, data=data_df, ax=axes[0,0])
axes[0,0].set_title("ViolinPlot for {} feature".format(feature))
axes[0,0].set_xlabel("Class Labels")
axes[0,0].set_ylabel(feature)

sns.boxplot(x="Label", y=feature, data=data_df, ax=axes[0,1])
axes[0,1].set_title("Boxplot for {} feature".format(feature))
axes[0,1].set_xlabel("Class Labels")
axes[0,1].set_ylabel(feature)

sns.violinplot(x="Label", y=feature, data=data_df[data_df.vr_source_avg_pkt<1e4], ax=axes[1,0])
axes[1,0].set_title("ViolinPlot for {} feature".format(feature))
axes[1,0].set_xlabel("Class Labels")
axes[1,0].set_ylabel(feature)

sns.boxplot(x="Label", y=feature, data=data_df[data_df.vr_source_avg_pkt<1e4], ax=axes[1,1])
axes[1,1].set_title("Boxplot for {} feature".format(feature))
axes[1,1].set_xlabel("Class Labels")
axes[1,1].set_ylabel(feature)


plt.show()
```

<li> First row plots are without trimming the values while second row, the feature value is capped at 1e4 </li>
<li> It can be seen with above plots that source average packet size is a useful feature and with this feature, normal datapoints can be identified with high vr_source_avg_pkt value. </li>

```python
fig, axes = plt.subplots(1, 1, figsize=(12, 6))
fig.suptitle("Attack types vs {}".format(feature))

sns.boxplot(x="attack_cat", y=feature, data=data_df[data_df.attack_cat != 'normal'])
plt.xlabel("Attack Types")
plt.ylabel(feature)
plt.show()
```

<li> As it can be seen that this feature is able to distinguish some points of Exploits, DoS, Generic & Fuzzers attack categories which have high value of vr_source_avg_pkt. </li>
